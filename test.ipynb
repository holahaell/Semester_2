```python
with open("data.txt","r") as file:
    read = file.readlines()
    data = []
    
    for line in read:
        temp = re.split(r"\s{2,}|\n",line)
        data.append(temp[0:4])
# ini baca file
```

```python
df = pd.DataFrame(data, columns =['nama', 'nim', 'asal', 'hobi'])
df
# ini create tabel
```

```python
df = df.append({'name':'', 'nim':'', 'asal':''}, ignore_index=True)
# ini insert
```

```python
search_by = "nama" 
value = ""
index = df[df[search_by].str.lower().str.contains(value)].index
# ini search
```

```python
df.drop(index, inplace=True)
# ini delete
```

```python

df.loc[index,["nama"]] = ["FARAH RAHIL NURIN SHABRINA"]
# update
```

```python
with open('dataex.txt', 'w') as f: df.to_string(f, col_space=10, header=False, index=False)
# save to file
```
